```python
luke_skill = ('python', 'sql', 'statistics', 'tableau')
print(id(luke_skill))

new_skill = ('excel', 'power bi')

lukes_skills = luke_skill + new_skill

lukes_skills
print(id(lukes_skills))
```

```python
luke_skill = ['python', 'sql', 'statistics', 'tableau']
print(id(luke_skill))

new_skill = ['excel', 'power bi']

luke_skill += new_skill

luke_skill
print(id(luke_skill))
```

#User-Defined Function

```python
#Calculate total salary of base_salary = (1 + bonus_rate)

def calculate_salary(base_salary, bonus_rate=.1):

  total_salary = base_salary * (1 + bonus_rate)
  return total_salary
```

```python
base_salary = 100000
bonus_rate = 0.1

calculate_salary(100000)
```

# Data Filtering

```python
job_data = [
    {
        "job_title": "Data Scientist",
        "job_skill": ["Python", "Machine Learning", "Statistics", "Data Analysis"],
        "remote": True
    },
    {
        "job_title": "Data Analyst",
        "job_skill": ["SQL", "Excel", "Data Visualization", "Python"],
        "remote": False
    },
    {
        "job_title": "Machine Learning Engineer",
        "job_skill": ["Python", "TensorFlow", "Deep Learning", "Algorithms"],
        "remote": True
    },
    {
        "job_title": "Business Intelligence Analyst",
        "job_skill": ["SQL", "Power BI", "Tableau", "Data Warehousing"],
        "remote": False
    },
    {
        "job_title": "Data Engineer",
        "job_skill": ["Python", "SQL", "Big Data", "ETL Processes"],
        "remote": True
    }
]
```

```python
list(filter(lambda job: job['remote'] and "Python" in job['job_skill'], job_data))
```
